```python
import matplotlib.pyplot as plt

# Define the points
x1, y1 = 1, 1
x2, y2 = 9, 3

# Calculate the slope
slope = (y2 - y1) / (x2 - x1)

# Generate x and y values for the line
x_values = [x1, x2]
y_values = [y1, y2]

# Plot the line
plt.plot(x_values, y_values, label=f'Line with slope {slope:.2f}')

# Plot the points
plt.scatter([x1, x2], [y1, y2], color='red', label='Points')
plt.text(x1, y1, f'({x1}, {y1})', ha='right', va='bottom')
plt.text(x2, y2, f'({x2}, {y2})', ha='right', va='bottom')

# Add labels and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Passing Through (1, 1) and (9, 3)')
plt.legend()

# Show the plot
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```
